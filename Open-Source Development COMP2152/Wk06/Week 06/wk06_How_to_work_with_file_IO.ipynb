```
#How to write and read a list of strings
```

```
#How to write the items in a list to a file
members = ["John Cleese", "Eric Idle"]
with open("/content/drive/MyDrive/PythonClass/members.txt", "w") as file:
    for m in members:
        file.write(m + "\n") 
```

```
#How to read the lines in a file into a list
members = []
with open("/content/drive/MyDrive/PythonClass/members.txt", "r") as file:
    for line in file:
        line = line.replace("\n", "") # removes new line
        members.append(line)
print(members)
```

```
#How to write and read a list of numbers
```

```
#How to write the items in a list to a file
years = [1975, 1979, 1983]
with open("/content/drive/MyDrive/PythonClass/years.txt", "w") as years_file:
    for year in years:
        years_file.write(str(year) + "\n")  # converts int to str
```

```
#How to read the items in a list from a file
years = []
with open("/content/drive/MyDrive/PythonClass/years.txt") as file:
    for line in file:
        line = line.replace("\n", "")
        years.append(line)
print(years)
```

```
#How to read the items in a list from a file
years = []
with open("/content/drive/MyDrive/PythonClass/years.txt") as file:
    for line in file:
        line = line.replace("\n", "")
        years.append(int(line))    # converts str to int
print(years)
```

```
#CSV time!!!
#Creating a CSV file.
```

```
movies = [["Monty Python and the Holy Grail", 1975],
          ["Cat on a Hot Tin Roof", 1958],
          ["On the Waterfront", 1954]]
```

```
#How to import the csv module
import csv
```

```
#How to write the list to a CSV file
with open("/content/drive/MyDrive/PythonClass/movies.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(movies)
```

```
#The reader() function of the csv module
```

```
#How to read data from a CSV file
with open("/content/drive/MyDrive/PythonClass/movies.csv", newline="") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row[0] + " (" + str(row[1]) + ")")
```

```
#How to save a binary file
```

```
movies = [["Monty Python and the Holy Grail", 1975],
          ["Cat on a Hot Tin Roof", 1958],
          ["On the Waterfront", 1954]]
```

```
# How to import the pickle module
import pickle
```

```
# How to write an object to a binary file
with open("/content/drive/MyDrive/PythonClass/movies.bin", "wb") as file:  # write binary
    pickle.dump(movies, file)
```

```
#How to open a binary file
```

```
#How to read an object from a binary file
```

```
with open("/content/drive/MyDrive/PythonClass/movies.bin", "rb") as file: # read binary
    movie_list = pickle.load(file)
    print(movie_list)
```
