```
scores = [70, 80, 90, 100]
total = 0
for score in scores:
    total += score
print(total)
```

```
# With a while loop
```

```
scores = [70, 80, 90, 100]
total = 0
i = 0
while i < len(scores):
    total += scores[i]
    i += 1
print(total)
```

```
# How to make a shallow copy of a list
list_one = [1, 2, 3, 4, 5]
list_two = list_one
list_two[1] = 4
print(list_one)		# [1, 4, 3, 4, 5]
print(list_two)		# [1, 4, 3, 4, 5]
```

```
list_one[0] = 10
print(list_one)		# [10, 4, 3, 4, 5]
print(list_two)	
```

```
# How to make a deep copy of a list
import copy
list_one = [1, 2, 3, 4, 5]
list_two = copy.deepcopy(list_one)
list_two[1] = 4
print(list_one)		# [1, 2, 3, 4, 5]
print(list_two)		# [1, 4, 3, 4, 5]
```
